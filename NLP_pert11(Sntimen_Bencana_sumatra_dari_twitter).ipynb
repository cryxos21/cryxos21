Langkah 5: Cek Hasil Data
Setelah proses selesai, data tersimpan dalam file CSV. Mari kita buka isinya menggunakan Pandas untuk memastikan data berhasil terambil.

```python
import pandas as pd

# Membaca file hasil crawling
# Sesuaikan path foldernya, biasanya ada di folder tweets-data
path_file = "/content/tweets-data/data_tweet_bencana.csv"
df = pd.read_csv(path_file, sep=",")

# Menampilkan informasi data
df.info()

# Menampilkan 10 data teratas
df.head(10)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# 1. Gabungkan seluruh teks tweet
all_text = ' '.join(df['full_text'].astype(str))

# 2. DAFTAR KATA SAMPAH (STOPWORDS) YANG MAU DIHAPUS
# Kamu bisa tambah kata lain di dalam kurung kurawal {} ini
stopwords_lokal = {
    'di', 'yang', 'dan', 'dari', 'ke', 'ini', 'itu', 'untuk',
    'pada', 'adalah', 'dengan', 'saya', 't', 'co', 'https', 'http',
    'rt', 'rw', 'nan', 'terkait', 'bisa', 'akan', 'sudah'
}

# 3. Buat Word Cloud dengan Filter
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='black',
    colormap='magma',
    stopwords=stopwords_lokal, # <--- Ini kuncinya!
    min_font_size=10
).generate(all_text)

# 4. Tampilkan Hasil
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Topik Utama (Versi Bersih)', fontsize=15, color='white', pad=20)
plt.show()
```

```python
#Timeline Aktivitas (Grafik Garis/Batang)
import seaborn as sns

# 1. Pastikan kolom 'created_at' formatnya tanggal yg benar
df['created_at'] = pd.to_datetime(df['created_at'])

# 2. Ekstrak Tanggal saja (untuk sumbu X)
df['Tanggal'] = df['created_at'].dt.date

# 3. Hitung jumlah tweet per hari
daily_counts = df['Tanggal'].value_counts().sort_index()

# 4. Plotting
plt.figure(figsize=(10, 6))
# Pakai style darkgrid biar kelihatan profesional
sns.set_style("darkgrid")

# Buat Bar Chart
ax = sns.barplot(x=daily_counts.index, y=daily_counts.values, palette="Reds_d")

plt.title('Lonjakan Tweet Bencana Sumatera (1-7 Des 2025)', fontsize=16, fontweight='bold')
plt.xlabel('Tanggal', fontsize=12)
plt.ylabel('Jumlah Tweet', fontsize=12)
plt.xticks(rotation=45) # Miringkan tanggal biar ga tubrukan
plt.show()
```

```python
#Top 5 Tweet Paling Viral (Tabel / Screenshot)
# Urutkan data berdasarkan jumlah retweet terbanyak
top_viral = df.sort_values(by='retweet_count', ascending=False).head(5)

# Pilih kolom yg mau ditampilkan aja
tabel_viral = top_viral[['created_at', 'username', 'full_text', 'retweet_count', 'favorite_count']]

# Tampilkan
print("=== TOP 5 TWEET PALING VIRAL ===")
display(tabel_viral)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import textwrap

# 1. Ambil Top 5 Data berdasarkan Retweet
top5 = df.sort_values(by='retweet_count', ascending=False).head(5)

# 2. Fungsi biar tulisan tweet-nya gak lurus panjang ke kanan (di-enter otomatis/wrapping)
def wrap_text(text, width=50):
    # Potong teks jadi maksimal 50 karakter per baris, dan ambil 2 baris pertama aja biar rapi
    wrapped = textwrap.wrap(text, width)
    return "\n".join(wrapped[:2]) + "..." if len(wrapped) > 2 else "\n".join(wrapped)

# Terapkan fungsi di atas ke kolom teks
top5['label_grafik'] = top5['full_text'].apply(lambda x: wrap_text(x, 40))

# 3. Mulai Menggambar Grafik
plt.figure(figsize=(12, 7)) # Ukuran gambar (Lebar, Tinggi)
sns.set_style("whitegrid")  # Background putih garis-garis biar bersih

# Buat Horizontal Bar Plot
# palette='magma' biar warnanya senada sama Word Cloud kamu tadi
ax = sns.barplot(x='retweet_count', y='label_grafik', data=top5, palette='magma')

# 4. Mempercantik Tampilan
plt.title('🏆 Top 5 Tweet Paling "Pecah" (Viral)', fontsize=20, fontweight='bold', pad=20)
plt.xlabel('Jumlah Retweet', fontsize=14)
plt.ylabel('', fontsize=12) # Label Y dikosongin aja biar bersih
plt.yticks(fontsize=11)     # Ukuran tulisan isi tweet

# Menambahkan Angka Persis di ujung batang (biar orang tau angkanya)
for container in ax.containers:
    ax.bar_label(container, padding=5, fontsize=12, fontweight='bold', color='black')

# Hapus garis kotak pinggir biar kelihatan modern
sns.despine(left=True, bottom=True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Ambil Top 5 Data (Pastikan urutannya sama kayak tadi)
top5 = df.sort_values(by='retweet_count', ascending=False).head(5)

# 2. KITA GANTI TEKS PANJANG JADI LABEL SINGKAT (MANUAL)
# Ini saya buatkan berdasarkan urutan data tweet kamu yang tadi:
# No 1. Tweet Relawan (Solar Panel)
# No 2. Tweet BNPB (Data Resmi)
# No 3. Tweet Fans K-Pop (Donasi Mark Lee)
# No 4. Tweet Netizen (Kritik Gubernur)
# No 5. Tweet PBB/WMO (Sorotan Internasional)

label_baru = [


    "Status Darurat: Aceh & Sumut",
    "Bantuan Internasional: Relief Sumatera",
    "Update Data: Jumlah Korban",
    "Laporan Donasi: Dana Terkumpul",
    "Solidaritas: Bantuan Mahasiswa"
]

# Masukkan label baru ke data
top5['Label_Singkat'] = label_baru

# 3. Mulai Menggambar Grafik
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

# Plot pakai Label Singkat
ax = sns.barplot(x='retweet_count', y='Label_Singkat', data=top5, palette='magma')

# 4. Hiasan biar cantik
plt.title('Siapa Paling Viral? (Top 5 Tweet)', fontsize=18, fontweight='bold', pad=20)
plt.xlabel('Jumlah Retweet', fontsize=12)
plt.ylabel('', fontsize=12)
plt.yticks(fontsize=12, fontweight='bold') # Huruf label ditebalkan biar jelas

# Kasih angka di ujung batang
for container in ax.containers:
    ax.bar_label(container, padding=5, fontsize=12, fontweight='bold')

sns.despine(left=True, bottom=True)
plt.tight_layout()
plt.show()
```
